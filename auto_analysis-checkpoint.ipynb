```python
import pandas as pd

# Load the Stata data file
file_path = 'auto.dta'
auto_data = pd.read_stata(file_path)

# Display the first few rows of the dataframe to understand its structure
auto_data.head()
```

### Step 1: View the attributes of the data

```python
# View the attributes of the data
auto_data.info()
```

### Step 2: Univariate Data Exploratory Analysis - price

```python
# Univariate Data Exploratory Analysis - price
price_stats = auto_data['price'].describe()
price_stats
```

```python
import matplotlib.pyplot as plt

# histogram
plt.figure(figsize=(10, 6))
plt.hist(auto_data['price'], bins=20, edgecolor='black')
plt.title('Distribution of Car Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.savefig('price_distribution.png')
plt.show()
```

```python
# Box plot
plt.figure(figsize=(10, 6))
plt.boxplot(auto_data['price'], vert=False)
plt.title('Boxplot of Car Prices')
plt.xlabel('Price')
plt.grid(True)
plt.savefig('price_boxplot.png')
plt.show()
```

### Step 3: Explore the relationship between variables and other variables

```python
# the relationship between variables and other variables
mpg_price_pivot = auto_data.pivot_table(values='price', index='mpg', aggfunc='mean')
mpg_price_pivot
```

```python
# 
weight_price_pivot = auto_data.pivot_table(values='price', index='weight', aggfunc='mean')
weight_price_pivot
```

```python
# The relationship between price and whether the car is imported or not
foreign_price_pivot = auto_data.pivot_table(values='price', index='foreign', aggfunc='mean')
foreign_price_pivot
```

```python
# Scatter chart - Price and mileage
plt.figure(figsize=(10, 6))
plt.scatter(auto_data['mpg'], auto_data['price'], alpha=0.7)
plt.title('Price vs. Mileage')
plt.xlabel('Mileage (mpg)')
plt.ylabel('Price')
plt.grid(True)
plt.savefig('price_vs_mileage.png')
plt.show()
```

```python
# Scatter plot - Price and weight
plt.figure(figsize=(10, 6))
plt.scatter(auto_data['weight'], auto_data['price'], alpha=0.7)
plt.title('Price vs. Weight')
plt.xlabel('Weight')
plt.ylabel('Price')
plt.grid(True)
plt.savefig('price_vs_weight.png')
plt.show()
```

```python
# Bar chart - Price with or without imported car
plt.figure(figsize=(10, 6))
auto_data.groupby('foreign')['price'].mean().plot(kind='bar', color=['blue', 'orange'])
plt.title('Average Price by Car Origin')
plt.xlabel('Car Origin')
plt.ylabel('Average Price')
plt.grid(True)
plt.savefig('price_by_origin.png')
plt.show()
```
